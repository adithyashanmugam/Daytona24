```python
import pandas as pd
```

```python
concentration = pd.read_csv('./data/concentrations.csv')
labels = pd.read_csv('./data/yellow_labels.csv')
pit = pd.read_csv('./data/avg_time_since_pit.csv')
weather = pd.read_csv('./data/weather.csv', sep=';')
```

```python
max_concentration = pd.DataFrame(columns=['Time', 'Max Concentration'])
temp_max = 0
for index, row in concentration.iterrows():
    sector = row[1]
    if sector == 1:
        temp_max = 0
    if row[2] > temp_max:
        temp_max = row[2]
    if sector == 13:
        max_concentration.loc[index] = [row[0], temp_max]
```

```python
weather_mod = pd.DataFrame(columns=['Time', 'Track Temp', 'Wind Speed','Rain'])
start_time = weather['TIME_UTC_SECONDS'].iloc[0]

with open('./data/weather_mod.csv', 'w') as f:
    f.write('Time,Track Temp,Wind Speed, Rain\n')
    for index, row in weather.iterrows():
            f.write("%s,%s,%s,%s\n"%(row[0]-start_time, row[3], row[6], row[8]))
            f.write("%s,%s,%s,%s\n"%(row[0]-start_time+20, row[3], row[6], row[8]))
            f.write("%s,%s,%s,%s\n"%(row[0]-start_time+40, row[3], row[6], row[8]))
```

```python
agg = max_concentration.merge(pit, how='inner', on='Time')
agg = agg.merge(labels, how='inner', on='Time')
agg = weather_mod.merge(agg, how='inner', on='Time')
agg
```

```python
agg.to_csv('./data/agg_data.csv', index=False)
```
