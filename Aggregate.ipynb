```python
import pandas as pd
```

```python
concentration = pd.read_csv('./data/concentrations.csv')
labels = pd.read_csv('./data/yellow_labels.csv')
concentration.iloc[0]
```

```python
max_concentration = pd.DataFrame(columns=['Time', 'Max Concentration'])
temp_max = 0
for index, row in concentration.iterrows():
    sector = row[1]
    if sector == 1:
        temp_max = 0
    if row[2] > temp_max:
        temp_max = row[2]
    if sector == 13:
        max_concentration.loc[index]{'Time' : row[0], 'Max Concentration': temp_max}, ignore_index=True)
max_concentration = pd.DataFrame.from_dict(max_concentration)
agg = max_concentration.merge(labels, how='inner', on='Time')
agg
```

```python
agg.to_csv('./data/agg_data.csv', index=False)
```
