```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/agg_data.csv')
X = df.iloc[:, 1:-4]
y = df.iloc[:, -1]
X
```

```python
clf = LogisticRegression(random_state=0).fit(X, y)
yg = clf.predict_proba(X)[:, 1]
```

```python
xg = list(df.Time)
```

```python
plt.plot(xg, yg)
t = zip(xg, yg)
n = [a[1] for a in t if a[0] == 70540.0]
n
```
