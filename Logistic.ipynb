```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./data/agg_data.csv')
# df = df.iloc[94:, :]
pit = df.iloc[:, 2]
# temp = df.iloc[:, 3]
# time = df.iloc[:, 0]
label = df.iloc[:, -1]

z = list(zip(pit, label))
z = sorted(z, key=lambda t: t[0])

X = [t[0] for t in z]
X = np.array(X).reshape(-1, 1)
y = [t[1] for t in z]
```

```python
clf = LogisticRegression(random_state=0).fit(X, y)
yg = clf.predict_proba(X)[:, 1]
```
