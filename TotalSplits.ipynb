```python
import pandas as pd
from datetime import datetime, timedelta, date
from collections import defaultdict
import json
```

```python
df = pd.read_csv('./data/2020_jan_mod.csv')
```

```python
flag_times = []
green_flag = True
for index, row in df.iterrows():
    if row.Flag == 'Green':
        green_flag = True
    elif green_flag and row.Flag == 'Yellow':
        flag_times.append(row['Session Time'])
        green_flag = False
flag_times[0]
```

```python
first_flag = datetime.strptime(flag_times[0], '%Y-%m-%d %H:%M:%S.%f')
starting_date = datetime(1900, 1, 1)
flag_seconds = (first_flag - starting_date).total_seconds()
flag_seconds
```
